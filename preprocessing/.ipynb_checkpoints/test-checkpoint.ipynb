```python
# -*- coding: utf-8 -*-
# @Time    : 2018/4/30 13:19
# @Author  : Intgrp

import numpy as np
import matplotlib.pyplot as plt
import os
from skimage import color
from skimage import exposure,data
from skimage import data,filters

import drop_plastic
import dataRead as dataRead
```

```python
#简单读取水果类别和各类水果的图片名称
fruitsName, fruitsImages = dataRead.file_name()
```

```python
img = plt.imread(fruitsImages[0][0])
plt.subplot(221)
plt.imshow(img), plt.title('原图')

plt.subplot(222)
rst = color.rgb2hsv(img)
rst *= 255
rst.astype(np.uint8)
plt.imshow(rst[:,:,0],cmap='gray'), plt.title('HSV图的H分量')
plt.subplot(223)
plt.imshow(rst[:,:,1],cmap='gray'), plt.title('HSV图的S分量')
plt.subplot(224)
plt.imshow(rst[:,:,2],cmap='gray'), plt.title('HSV图的V分量')
plt.show()
```

```python
imgpath = fruitsImages[0][0]
img=plt.imread(imgpath)
mask = drop_plastic.fruit_collect(img)
```

```python
img.shape
```

```python
import copy
[n,m,c]=img.shape
out = copy.deepcopy(img)
```

```python
for i in range(n):
    for j in range(m):
        out[i][j][0] = out[i][j][0]*mask[i][j]
        out[i][j][1] = out[i][j][1]*mask[i][j]
        out[i][j][2] = out[i][j][2]*mask[i][j]
plt.imshow(out)
plt.show()
```
